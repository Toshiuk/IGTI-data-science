Linear Regression

```
prod = np.array([877.952, 1125.295, 1317.822, 2420.78, 2625.742, 2686.193, 2916.534, 3219.187, 3368.42, 3375.228, 3518.441, 3552.19, 3617.581, 3618.559, 3800.64, 4333.917, 4809.679, 4814.012, 4970.057, 5274.943])
cons = np.array([604.685, 682.931, 957.309, 1625.642, 2446.6, 1745.406, 2445.433, 2898.25, 1875.934, 2157.475, 2394.205, 2060.597, 2096.973, 2419.206, 2427.194, 3226.001, 2925.945, 2723.541, 3105.396, 3996.346])

regr = linear_model.LinearRegression()
regr.fit(prod.reshape(-1, 1), cons)
y = regr.predict(prod.reshape(-1, 1))

print('Coefficients: \n', regr.coef_)
print('Intercept: \n', regr.intercept_)

plt.scatter(prod, cons, color='black')
plt.plot(prod, y, color='blue', linewidth=3)

plt.show()
```
